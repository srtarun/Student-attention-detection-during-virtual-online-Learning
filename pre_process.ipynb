```python
import os
import cv2
```

```python
#create video frames

phases = ['Validation'] #'Train', 'Test', 

for phase in phases:
    path = os.path.join('DAiSEE/DataSet', phase)
    # print(path)
    subjects = os.listdir(path)
    # print(subjects)

    for subject in subjects:
        print(phase, subject, flush=True)
        videos = os.listdir(os.path.join(path,subject))
        for video in videos:
            videoPath = os.path.join(path,subject, video, os.listdir(os.path.join(path,subject,video))[0])
            # print(videoPath)
            videoPathFrames = '/'.join(videoPath.split('.')[0].split('/')[:-1]).replace(phase, phase + 'Frames')
            # print(videoPathFrames)
            os.makedirs(videoPathFrames, exist_ok=True)

            capture = cv2.VideoCapture(videoPath)
            frame_count = int(capture.get(cv2.CAP_PROP_FRAME_COUNT))
            count = 0
            i = 0
            retaining = True
            while count < frame_count and retaining:
                retaining, frame = capture.read()
                if frame is None:
                    continue
                cv2.imwrite(filename=os.path.join(videoPathFrames, '{}.jpg'.format(str(i))), img=frame)
                i += 1
                count+=1
            capture.release()
```

```python
#create label files
import pandas as pd
import os

phases = ['Train', 'Test', 'Validation']
labels = pd.read_csv( os.path.join('DAiSEE/Labels/', 'AllLabels.csv'))

for phase in phases:
    path = os.path.join('DAiSEE/DataSet', phase)
    phaseLabels = labels = pd.read_csv( os.path.join('DAiSEE/Labels/', phase+'Labels.csv'))
    df = pd.DataFrame(columns=['path', 'label'])
    subjects = os.listdir(path)
    skipped_cnt = 0
    for subject in subjects:
        print(phase, subject, flush=True)
        videos = os.listdir(os.path.join(path,subject))
        for video in videos:
            videoPath = os.path.join(path,subject, video, os.listdir(os.path.join(path,subject,video))[0])
            clipId = videoPath.split('/')[-1]
            videoPathFrames = '/'.join(videoPath.split('.')[0].split('/')[:-1]).replace(phase, phase + 'Frames')
            # print(videoPathFrames)
            # print(clipId)
            engagement = labels[labels['ClipID']==clipId]['Engagement'].values
            # print(labels[labels['ClipID']==clipId])
            if len(engagement) == 0:
                # print("modifying df")
                # print("All labels eng:", engagement)
                engagement = phaseLabels[phaseLabels['ClipID']==clipId]['Engagement'].values
                # print(phaseLabels[phaseLabels['ClipID']==clipId])
                # print("Phase lables eng:", engagement)
            # print(engagement)
            if len(engagement) == 0:
                skipped_cnt += 1
                continue
            df.loc[len(df)] = {'path':videoPathFrames, 'label':engagement[0]}
    df.to_csv(phase+'.csv')
    print(skipped_cnt)
```
